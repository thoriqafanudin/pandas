```python
import numpy as np
np.random.seed(100)
baris=[]
for i in range(100):
    baris.append(i)
    
data=np.random.rand(100,5)
kolom=['A','B','C','D','E']
baru=pd.DataFrame(data,baris,kolom)
baru.head(5)
```

```python
baru.plot.area(alpha=0.35)
```

```python
data=np.random.rand(4,5)
baru2=pd.DataFrame(data,[0,1,2,3],kolom)
baru2.plot.bar()
```

```python
baru2.plot.bar(stacked=True)
```

```python
baru2['C'].plot.bar()
```

```python
baru.plot.scatter(x='A',y='B')
```

```python
baru.plot.scatter(x='A',y='B',c='E',cmap='Reds')
```

```python
baru.plot.scatter(x='A',y='B',s=baru.D*100,c='C',cmap='Reds')
```

```python
baru.plot.box()
```

```python
baru.plot.line(y='D',lw=1)
```

```python
baru.plot.hexbin(x='A',y='B',gridsize=10)
```
