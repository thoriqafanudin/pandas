```python
import pandas as pd
satu=pd.DataFrame({'A':['a0','a1','a2','a3','a4','a5'],
                  'B':['b0','b1','b2','b3','b4','b5'],
                  'C':['c0','c1','c2','c3','c4','c5'],
                  'D':['d0','d1','d2','d3','d4','d5']},
                 index=[0,1,2,3,4,5])

dua=pd.DataFrame({'A':['a6','a7','a8','a9','a10','a11'],
                  'B':['b6','b7','b8','b9','b10','b11'],
                  'C':['c6','c7','c8','c9','c10','c11'],
                  'D':['d6','d7','d8','d9','d10','d11']},
                 index=[6,7,8,9,10,11])

tiga=pd.DataFrame({'A':['a12','a13','a14','a15','a16','a17'],
                  'B':['b12','b13','b14','b15','b16','b17'],
                  'C':['c12','c13','c14','c15','c16','c17'],
                  'D':['d12','d13','d14','d15','d16','d17']},
                 index=[12,13,14,15,16,17])

satu
```

Concatenation

```python
pd.concat([satu,dua,tiga])
```

```python
bisnis=pd.DataFrame({'Perusahaan':['Pertamina','BRI','KFC','KFC','Indosat','Pertamina','BRI','Telkom'],
                    'Karyawan':['Ucup','Otong','Asep','Dudung','Ayu','Bambang','Firman','Agus'],
                    'Usia':[50,27,19,35,41,20,29,55]})

bisnis
```

```python
bisnis.groupby('Perusahaan')
```

```python
bisnis.groupby('Perusahaan').sum()
```

```python
bisnis.groupby('Perusahaan').count()
```

```python
bisnis.groupby('Perusahaan').mean()
```

```python
bisnis.groupby('Perusahaan').max()
```
