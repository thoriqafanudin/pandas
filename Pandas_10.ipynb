# Merge

```python
import pandas as pd
satu=pd.DataFrame({'key1':['K0','K0','K1','K2'],
                  'key2':['K0','K1','K0','K1'],
                  'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']})

dua=pd.DataFrame({'key1':['K0','K1','K1','K2'],
                  'key2':['K0','K0','K0','K0'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})

satu
```

```python
dua
```

```python
baru=pd.merge(satu,dua,how='left',on=['key1','key2'])
baru
```

```python
baru=pd.merge(satu,dua,how='right',on=['key1','key2'])
baru
```

```python
baru=pd.merge(satu,dua,how='outer',on=['key1','key2'])
baru
```

```python
bapak=pd.DataFrame({'A':['A0','A1','A2'],
                   'B':['B0','B1','B2']},
                  index=['K0','K1','K2'])

ibu=pd.DataFrame({'C':['C0','C1','C2'],
                   'D':['D0','D1','D2']},
                  index=['K0','K2','K3'])

bapak.join(ibu)
```

```python
bapak.join(ibu,how='right')
```

```python
bapak.join(ibu,how='inn')
```
