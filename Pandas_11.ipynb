```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
data = pd.read_csv('kepadatan.csv')


data['L'] = data['35-39 Laki-Laki'] + data['40-44 Laki-Laki'] + data['45-49 Laki-Laki'] + data['50-54 Laki-Laki'] + data['55-59 Laki-Laki']
data['P'] = data['35-39 Perempuan'] + data['40-44 Perempuan'] + data['45-49 Perempuan'] + data['50-54 Perempuan'] + data['55-59 Perempuan']

kep_seribu = [sum(data['L'][0:6].to_numpy()), sum(data['P'][0:6].to_numpy())]
jak_pusat = [sum(data['L'][6:50].to_numpy()), sum(data['P'][6:50].to_numpy())]
jak_utara = [sum(data['L'][50:81].to_numpy()), sum(data['P'][50:81].to_numpy())]
jak_barat = [sum(data['L'][81:137].to_numpy()), sum(data['P'][81:137].to_numpy())]
jak_selatan = [sum(data['L'][137:202].to_numpy()), sum(data['P'][137:202].to_numpy())]
jak_timur = [sum(data['L'][202:267].to_numpy()), sum(data['P'][202:267].to_numpy())]

warga = [kep_seribu, jak_pusat, jak_utara, jak_barat, jak_selatan, jak_timur]
warga

kota = ['KEPULAUAN SERIBU', 'JAKARTA PUSAT', 'JAKARTA UTARA', 'JAKARTA BARAT', 'JAKARTA SELATAN', 'JAKARTA TIMUR']

gender = ['Laki-laki', 'Perempuan']

gambar = pd.DataFrame(warga, kota, gender)

gambar.plot.bar()
```
